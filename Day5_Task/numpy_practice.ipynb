```python
array_1d = [11, 26, 31, 30, 28]

array_2d = [
    [8, 2, 7],
    [4, 9, 6]
]


sum_1d = sum(array_1d)
sum_2d = sum(sum(row) for row in array_2d)

product_1d = 1
for num in array_1d:
    product_1d *= num

product_2d = 1
for row in array_2d:
    for num in row:
        product_2d *= num

max_1d = max(array_1d)
max_2d = max(max(row) for row in array_2d)

print("1D Array:", array_1d)
print("Sum of 1D Array:", sum_1d)
print("Product of 1D Array:", product_1d)
print("Maximum in 1D Array:", max_1d)

print("\n2D Array:")
for row in array_2d:
    print(row)

print("Sum of 2D Array:", sum_2d)
print("Product of 2D Array:", product_2d)
print("Maximum in 2D Array:", max_2d)
```
