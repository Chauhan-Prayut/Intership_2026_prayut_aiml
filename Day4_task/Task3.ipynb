```python
A = [
    [1, 2],
    [3, 4]
]

B = [
    [5, 6],
    [7, 8]
]

add_2d = [
    [A[i][j] + B[i][j] for j in range(len(A[0]))]
    for i in range(len(A))
]

mul_2d = [
    [
        A[i][0] * B[0][j] + A[i][1] * B[1][j]
        for j in range(len(B[0]))
    ]
    for i in range(len(A))
]

max_2d = max(max(row) for row in A)

print("\n2D Matrix A:")
for row in A:
    print(row)

print("\n2D Matrix B:")
for row in B:
    print(row)

print("\n2D Addition:")
for row in add_2d:
    print(row)

print("\n2D Multiplication:")
for row in mul_2d:
    print(row)

print("\n2D Maximum Value:", max_2d)
```
