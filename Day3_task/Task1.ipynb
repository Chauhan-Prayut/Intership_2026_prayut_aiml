```python
students = [
    "Prayut Chauhan",
    "Rushabh Shah",
    "Meet Vaja",
    "Jalpan Gohil",
    "Karan Patel"
]

for student in students:
    first_name, last_name = student.split()
    print("First Name:", first_name)
    print("Last Name:", last_name)
    print()

print("Total Count:", len(students))
```
