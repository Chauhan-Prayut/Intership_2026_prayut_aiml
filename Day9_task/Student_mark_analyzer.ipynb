```python
import pandas as pd

data = {
    "Student Name": ["Prayut", "Purva", "Krish", "Suraj", "Ronak"],
    "Maths": [85, 78, 92, 88, 76],
    "Science": [90, 82, 89, 91, 80],
    "English": [88, 75, 94, 86, 79]
}

df = pd.DataFrame(data)

df["Average Marks"] = df[["Maths", "Science", "English"]].mean(axis=1)

highest_marks = df[["Maths", "Science", "English"]].max().max()
lowest_marks = df[["Maths", "Science", "English"]].min().min()

top_student = df.loc[df["Average Marks"].idxmax(), "Student Name"]

print("Dataset:")
print(df)
 
print("\nAnalysis Results:")

print("Average Marks per Student:")
print(df[["Student Name", "Average Marks"]])

print("\nHighest Marks:", highest_marks)
print("Lowest Marks:", lowest_marks)

print("\nTop-Scoring Student:", top_student)
```
